```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

def getScores(actual,pred):
    mse = mean_squared_error(actual,pred)
    rmse = np.sqrt(mse)
    mae=mean_absolute_error(actual,pred)
    r2=r2_score(actual,pred)
    result={'mse':mse,'rmse':rmse,'mae':mae,'r2':r2}
    return result
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
maxFeat = ['auto','sqrt','log2']
```

```python
results = []
for feature in maxFeat:
    for depth in range(1,7):
        decReg =  DecisionTreeRegressor(max_depth=depth,max_features=feature)
        decReg.fit(xtrain,ytrain)
        result = getScores(ytest,decReg.predict(xtest))
        result['depth'] = depth
        result['feature'] = feature
        results.append(result)
```

```python
results
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
results = []
for feature in maxFeat:
    for depth in range(1,8):
        randReg =  RandomForestRegressor(max_depth=depth,max_features=feature)
        randReg.fit(xtrain,ytrain)
        result = getScores(ytest,randReg.predict(xtest))
        result['depth'] = depth
        result['feature'] = feature
        results.append(result)
```
